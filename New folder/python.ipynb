```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
df_2021 = pd.read_csv('2021.csv')
df_2022 = pd.read_csv('2022.csv')
df_2023 = pd.read_csv('2023.csv')
```

```python
df_2021.columns = df_2021.columns.str.replace(r'\W+', '_', regex=True).str.replace(r'_+', '_', regex=True).str.strip('_').str.lower()
df_2022.columns = df_2022.columns.str.replace(r'\W+', '_', regex=True).str.replace(r'_+', '_', regex=True).str.strip('_').str.lower()
df_2023.columns = df_2023.columns.str.replace(r'\W+', '_', regex=True).str.replace(r'_+', '_', regex=True).str.strip('_').str.lower()
```

```python
distinct_companies = set()
distinct_companies.update(set(df_2021['company_name']))
distinct_companies.update(set(df_2022['company_name']))
distinct_companies.update(set(df_2023['company_name']))
output_df = pd.DataFrame({'Companies': list(distinct_companies)})
```

```python
df2021 = pd.DataFrame(set(df_2021['company_name'].unique()),columns=['company_name'])
df2022 = pd.DataFrame(set(df_2022['company_name'].unique()),columns=['company_name'])
df2023 = pd.DataFrame(set(df_2023['company_name'].unique()),columns=['company_name'])
```

```python
output_df = pd.merge(output_df,df2021['company_name'], how='left', left_on='Companies', right_on='company_name')
output_df.rename(columns={'company_name':'2021'},inplace=True)
output_df = pd.merge(output_df,df2022['company_name'], how='left', left_on='Companies', right_on='company_name')
output_df.rename(columns={'company_name':'2022'},inplace=True)
output_df = pd.merge(output_df,df2023['company_name'], how='left', left_on='Companies', right_on='company_name')
output_df.rename(columns={'company_name':'2023'},inplace=True)
output_df
```

```python
output_df['2021'] = output_df['2021'].apply(lambda x: 'YES' if isinstance(x, str) else 'NO' if pd.isna(x) else x)
output_df['2022'] = output_df['2022'].apply(lambda x: 'YES' if isinstance(x, str) else 'NO' if pd.isna(x) else x)
output_df['2023'] = output_df['2023'].apply(lambda x: 'YES' if isinstance(x, str) else 'NO' if pd.isna(x) else x)
output_df.to_excel('output.xlsx',index=False)
```

```python
def anulizer(df):
    apdaf = 0
    redaf = 0
    odate = datetime.strptime(df[2], '%d-%m-%Y')
    if isinstance(df[4], str):
        if len(df[4]) > 4:
            adate = datetime.strptime(df[4], '%d-%m-%Y')
            apdaf = 1
            
    if isinstance(df[6], str):
        if len(df[6]) > 4:
            rdate = datetime.strptime(df[6], '%d-%m-%Y')
            redaf = 1
            
    if apdaf == 0 and redaf == 0:
        return
    
    if apdaf == 1 and redaf == 0:
        date_diff = odate - adate
        date_diff = date_diff.days 
        if date_diff > 0:
            l = [(df.iloc[1] * 365) / date_diff, (df.iloc[3] * 365) / date_diff, (df.iloc[5] * 365) / date_diff]
            return l
    
    if apdaf == 0 and redaf == 1:
        date_diff = (odate - rdate)
        date_diff = 365 - date_diff.days 
        if date_diff > 0:
            l = [(df.iloc[1] * 365) / date_diff, (df.iloc[3] * 365) / date_diff, (df.iloc[5] * 365) / date_diff]
            return l
            
    if apdaf == 1 and redaf == 1:
        date_diff = (rdate - adate)
        date_diff = date_diff.days 
        if date_diff > 0:
            l = [(df.iloc[1] * 365) / date_diff, (df.iloc[3] * 365) / date_diff, (df.iloc[5] * 365) / date_diff]
            return l
```

```python
desiglist = {'CEO - Chief Executive Officer','CFO - Chief Financial Officer','COO - Chief Operating Officer','CIO - Chief Information Officer','Company Secretary'}
```

```python
for desig in desiglist:    
    # desig = 'Company Secretary'
    # desig = 'COO - Chief Operating Officer'
    # desig = 'CEO - Chief Executive Officer'
    
    df = pd.DataFrame()
    df['company_name'] = output_df['Companies']

    temp = df_2021[df_2021['designation']==desig][['company_name','salary_indian_rupee']].sort_values(by=['company_name', 'salary_indian_rupee'], ascending=False).drop_duplicates(subset='company_name',keep='first')
    reqcols = df_2021[['date','directors_sitting_fees_indian_rupee','appointment_date_date','total_remuneration_indian_rupee','resignation_date_date']]
    temp = pd.merge(temp,reqcols,how ='left',left_index=True,right_index=True).dropna(subset='salary_indian_rupee')

    for k, v in temp.iterrows():
        res = anulizer(v)
        if res is not None:
            if res[1] is not None:
                temp.at[k, 'directors_sitting_fees_indian_rupee'] = res[1]
            if res[2] is not None:
                temp.at[k, 'total_remuneration_indian_rupee'] = res[2]
            if res[0] is not None:
                temp.at[k, 'salary_indian_rupee'] = res[0]
    temp.drop(columns=['date','appointment_date_date','resignation_date_date'],inplace = True)
    temp.rename(columns={'salary_indian_rupee':'2021_salary',
                        'directors_sitting_fees_indian_rupee':'2021_directors_fees',
                        'total_remuneration_indian_rupee':'2021_total_remuneration'},inplace=True)
    df = pd.merge(df,temp,how = 'left',on='company_name')
    #------------------------------------------------------------------------------------------------------
    temp = df_2022[df_2022['director_designation']==desig][['company_name','salary']].sort_values(by=['company_name', 'salary'], ascending=False).drop_duplicates(subset='company_name',keep='first')
    reqcols = df_2022[['date','directors_sitting_fees','appointment_date','total_remuneration','resignation_date']]
    temp = pd.merge(temp,reqcols,how ='left',left_index=True,right_index=True).dropna(subset='salary')

    for k, v in temp.iterrows():
        res = anulizer(v)
        if res is not None:
            if res[1] is not None:
                temp.at[k, 'directors_sitting_fees'] = res[1]
            if res[2] is not None:
                temp.at[k, 'total_remuneration'] = res[2]
            if res[0] is not None:
                temp.at[k, 'salary'] = res[0]
    temp.drop(columns=['date','appointment_date','resignation_date'],inplace = True)
    temp.rename(columns={'salary':'2022_salary',
                        'directors_sitting_fees':'2022_directors_fees',
                        'total_remuneration':'2022_total_remuneration'},inplace=True)
    df = pd.merge(df,temp,how = 'left',on='company_name')
    #------------------------------------------------------------------------------------------------------
    temp = df_2023[df_2023['designation']==desig][['company_name','salary_indian_rupee']].sort_values(by=['company_name', 'salary_indian_rupee'], ascending=False).drop_duplicates(subset='company_name',keep='first')
    reqcols = df_2023[['date','directors_sitting_fees_indian_rupee','appointment_date_date','total_remuneration_indian_rupee','resignation_date_date']]
    temp = pd.merge(temp,reqcols,how ='left',left_index=True,right_index=True).dropna(subset='salary_indian_rupee')

    for k, v in temp.iterrows():
        res = anulizer(v)
        if res is not None:
            if res[1] is not None:
                temp.at[k, 'directors_sitting_fees_indian_rupee'] = res[1]
            if res[2] is not None:
                temp.at[k, 'total_remuneration_indian_rupee'] = res[2]
            if res[0] is not None:
                temp.at[k, 'salary_indian_rupee'] = res[0]
    temp.drop(columns=['date','appointment_date_date','resignation_date_date'],inplace = True)
    temp.rename(columns={'salary_indian_rupee':'2023_salary',
                        'directors_sitting_fees_indian_rupee':'2023_directors_fees',
                        'total_remuneration_indian_rupee':'2023_total_remuneration'},inplace=True)
    df = pd.merge(df,temp,how = 'left',on='company_name')
    #------------------------------------------------------------------------------------------------------

    df['change_2021-2022'] = ((df['2022_salary'] - df['2021_salary']) / df['2021_salary']) * 100
    df['change_2022-2023'] = ((df['2023_salary'] - df['2022_salary']) / df['2022_salary']) * 100
    df['change_2021-2022'] = df['change_2021-2022'].apply(lambda x: f'{round(x):,.0f}%' if not pd.isna(x) and x != np.inf and x != -np.inf else x)
    df['change_2022-2023'] = df['change_2022-2023'].apply(lambda x: f'{round(x):,.0f}%' if not pd.isna(x) and x != np.inf and x != -np.inf else x)

    salary_columns = ['2021_salary', '2022_salary', '2023_salary']
    df.replace(0, np.nan, inplace=True)
    df = df[df[salary_columns].apply(lambda x: any(x >= 2000000), axis=1)]
    df.to_excel(desig+'.xlsx',index=False)
```
