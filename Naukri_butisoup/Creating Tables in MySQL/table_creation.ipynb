```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)
cursor = connection.cursor()

# Create table query
create_table_query = """
CREATE TABLE IF NOT EXISTS job (
    id INT PRIMARY KEY AUTO_INCREMENT,
    job_title VARCHAR(255),
    company_name VARCHAR(255),
    experience VARCHAR(255),
    salary VARCHAR(255),
    industry_id INT,
    department_id INT,
    role_category_id INT,
    role_id INT,
    salary_minimum FLOAT,
    salary_maximum FLOAT,
    experience_minimum INT,
    experience_maximum INT,
    posted_on DATE,
    extracted_on DATE,
    FOREIGN KEY (industry_id) REFERENCES industry(id),
    FOREIGN KEY (department_id) REFERENCES department(id),
    FOREIGN KEY (role_category_id) REFERENCES role_category(id),
    FOREIGN KEY (role_id) REFERENCES role(id)
);
"""

cursor.execute(create_table_query)

connection.commit()
cursor.close()
connection.close()
```

```python
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)
cursor = connection.cursor()

# Add foreign key constraint to job_location table
alter_table_query = """
ALTER TABLE job_location
ADD CONSTRAINT fk_job_id_location
FOREIGN KEY (job_id) REFERENCES job(id)
"""
cursor.execute(alter_table_query)

# Add foreign key constraint to job_skill table
alter_table_query = """
ALTER TABLE job_skill
ADD CONSTRAINT fk_job_id_skill
FOREIGN KEY (job_id) REFERENCES job(id)
"""
cursor.execute(alter_table_query)

# Commit changes and close connection
connection.commit()
cursor.close()
connection.close()
```
