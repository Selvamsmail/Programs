```python
skillcol = [col for col in df.columns if 'skill' in col and 'comb' not in col]
```

```python
mostcomskills = {}
for inde,row in df[(df['role'].str.contains(r'business development executive \(bde\)',case=False))].iterrows():
    for skill in row[skillcol].tolist():
        if skill in mostcomskills:
            mostcomskills[skill] += 1
        else:
            mostcomskills[skill] = 1
```

```python
pd.DataFrame.from_dict(mostcomskills, orient='index',columns=['count']).sort_values(by='count',ascending=False).head(30)
```
