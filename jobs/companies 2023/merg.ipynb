```python
import pandas as pd
```

```python
df_2021 = pd.read_excel('data_2021.xlsx')
df_2022 = pd.read_excel('data_2022.xlsx')
df_2023 = pd.read_excel('data_2023.xlsx')
```

```python
distinct_companies = set()
distinct_companies.update(set(df_2021['company_name']))
distinct_companies.update(set(df_2022['company_name']))
distinct_companies.update(set(df_2023['Company Name']))
```

```python
output_df = pd.DataFrame({'Companies': list(distinct_companies)})
df2021 = pd.DataFrame({'company_name': list(set(df_2021['company_name']))})
df2022 = pd.DataFrame({'company_name': list(set(df_2022['company_name']))})
df2023 = pd.DataFrame({'company_name': list(set(df_2023['Company Name']))})
```

```python
output_df = pd.merge(output_df,df2021['company_name'], how='left', left_on='Companies', right_on='company_name')
output_df.rename(columns={'company_name':'2021'},inplace=True)

output_df = pd.merge(output_df,df2022['company_name'], how='left', left_on='Companies', right_on='company_name')
output_df.rename(columns={'company_name':'2022'},inplace=True)

output_df = pd.merge(output_df,df2023['company_name'], how='left', left_on='Companies', right_on='company_name')
output_df.rename(columns={'company_name':'2023'},inplace=True)
output_df
```

```python
output_df.columns
```

```python
output_df['2021'] = output_df['2021'].apply(lambda x: 'Yes' if pd.notna(x) else 'No')
output_df['2022'] = output_df['2022'].apply(lambda x: 'Yes' if pd.notna(x) else 'No')
output_df['2023'] = output_df['2023'].apply(lambda x: 'Yes' if pd.notna(x) else 'No')
```

```python
output_df
```

```python
output_df.to_excel('output.xlsx',index=False)
```
