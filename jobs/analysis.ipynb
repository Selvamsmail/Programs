```python
df['mapped_average_sal'] = (df['salmin'] + df['salmax']) / 2 
```

```python
df['current_date'] = pd.to_datetime(df['current_date'])

# Filter data for 'finan' or 'bank' in industry_type
filtered_df = df[(df['industry_type'].str.contains('finan|bank', case=False, na=False)) & (df['mapped_job_title'] != 'unclassified')]

# Sort the values by 'mapped_average_sal'
sorted_df = filtered_df.sort_values(by='mapped_average_sal', ascending=False)
```

```python
sorted_df.info()
```

```python
sorted_df[(sorted_df['minexp'] > 2) & (sorted_df['minexp'] > 10)]
```

```python
for expran in ['0-2','2-5','5-8','8-10','10-13','13-16']:
    min_exp = int(expran.split('-')[0])
    max_exp = int(expran.split('-')[1])
        
    plt.figure(figsize=(12, 8))

    tempdf = sorted_df[(sorted_df['minexp'] >= min_exp) & (sorted_df['minexp'] <= max_exp)]
    sns.lineplot(data=tempdf, x='current_date', y='mapped_average_sal', markers=True, errorbar=None)

    # Add labels and title
    plt.xlabel('Date')
    plt.ylabel('Average Salary')
    plt.title(f'Trend Chart of Average Salary in Finance/Banking for Experience {expran}')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)
    date_form = DateFormatter("%Y-%b-%d")  # Abbreviated month names
    plt.gca().xaxis.set_major_formatter(date_form)
    plt.xticks(rotation=45)

    # Display the plot
    plt.tight_layout()
    filename = f'average_salary_{expran}.png'
    plt.savefig(filename)
    plt.close()
    plt.show()
```
